```python
def read_annotated_sentences(embedding_name, actor='patient', variation='without_others_intent/k100_without_sentences_higher_than_median'):
    file_name = fm.filename_from_data_dir(
        f'output/{actor}/{variation}/{embedding_name}/annotated_sentences.csv'
    )

    return pd.read_csv(file_name)
```

```python
# pd.merge(df_bert['txt'], df_glove['txt'], how ='inner', on ='txt')

dfs = [df_bert, df_flair, df_glove, df_lasbe, df_use]

# pd.concat(dfs, axis=1, join='inner')


df_merged = dfs[0]
for df in dfs[1:]:       
    df_merged = pd.merge(df_merged['txt'], df['txt'], on='txt', how='inner')

df_merged
```

```python
def get_path_df(embedding_name, type='train'):
    return fm.filename_from_data_dir(
        f'output/neural_models/patient/{embedding_name}/data_{type}.csv'
        )
```

```python
df_trains = [
  pd.read_csv(get_path_df(embedding_name=embedding_name, type='train'))
  for embedding_name in EMBEDDING_MODELS_TRANSLATION.keys()
]
```

```python
df_train_merged = pd.concat(df_trains, ignore_index=True)

df_train_merged = df_train_merged.drop_duplicates(subset=['correct_txt'])

df_train_merged
```

```python
df_train_merged['intent'].value_counts()
```

```python
df_tests = [
  pd.read_csv(get_path_df(embedding_name=embedding_name, type='test'))
  for embedding_name in EMBEDDING_MODELS_TRANSLATION.keys()
]

df_test_merged = pd.concat(df_tests, ignore_index=True)

df_test_merged = df_test_merged.drop_duplicates(subset=['correct_txt'])

df_test_merged
```

```python
df_test_merged['intent'].value_counts()
```

```python
used_data = pd.concat([df_train_merged, df_test_merged], ignore_index=True).drop_duplicates(subset=['correct_txt'])
```

```python
annotated_sentences = pd.read_csv(fm.filename_from_data_dir('output/patient/annotated_sentences.csv'))

annotated_sentences
```

```python
untrained_sentences = annotated_sentences[
  ~annotated_sentences['txt'].isin(df_train_merged['correct_txt'])
  ]
```

```python
not_used_sentences = annotated_sentences[
  ~annotated_sentences['txt'].isin(used_data['correct_txt'])
  ]
```
