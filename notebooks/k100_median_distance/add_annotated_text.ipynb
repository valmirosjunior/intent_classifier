```python
dict_annotated_txt = pd.Series(df_annotated.annotated_txt.values,index=df_annotated.txt).to_dict()

len(dict_annotated_txt.keys())
```

```python
def add_annottated_text(model_name):
  file_name_input = DataHelper.generate_filename_from_data_dir(f'output/patient/k100_median_distance/{model_name}/classified_sentences.csv')
  file_name_output = DataHelper.generate_filename_from_data_dir(f'output/patient/k100_median_distance/{model_name}/annotated_sentences.csv')

  df = pd.read_csv(file_name_input)
  df['annotated_txt'] = df['txt'].map(dict_annotated_txt)

  count_nam = df.isnull().sum().sum()
  print(f'There are {count_nam} rows with nan values')

  df = df.dropna()
  df.to_csv(file_name_output, index=False)


  return df
```
