```python
from src.core.chart_helper import *
from src.embeddings.constants import  EMBEDDING_MODELS_TRANSLATION


dfs = [
  pd.read_csv(
    fm.filename_from_data_dir(
      f'output/patient/k100/{embedding}/clustering_metrics.csv'
    )
  )

  for embedding in EMBEDDING_MODELS_TRANSLATION.keys()
]

# variation_map = {
#   'all'         : 'All',
#   'not_median'  : '~ median',
#   'not_outliers': '~ outiliers',
# }

# intent_variations_map = {
#   'all_intents'   : '',
#   'without_others': '~(others) - ',
# }

# variation_map = {
#   'all'         : 'All Sentences',
#   'not_median'  : 'Threshold = Mediana',
#   'not_outliers': 'Whitout outiliers',
# }

# intent_variations_map = {
#   'all_intents'   : 'All intents - ',
#   'without_others': 'Without others intent - ',
# }

variation_map = {
  'all'         : 'All Sentences',
  'not_median'  : 'Threshold = Median',
  'not_outliers': 'Whitout Outiliers',
}

intent_variations_map = {
  'all_intents'   : '',
  'without_others': ' Without others intent',
}

variations_legend = {
  'all_intents-all': 'A',
  'all_intents-not_outliers': 'B',
  'all_intents-not_median': 'C',
  'without_others-all': 'D',
  'without_others-not_outliers': 'E',
  'without_others-not_median': 'F'
}

def get_key(row):
  # intent_variations = intent_variations_map[row['intents']]
  
  # variation = variation_map[row['variation']]
  
  # return  f"{variation}{intent_variations}"

  return variations_legend[f"{row['intents']}-{row['variation']}"]

df = pd.concat(dfs)
df['Embedding Model'] = df['embedding'].map(EMBEDDING_MODELS_TRANSLATION)
df['Data Variation'] = df.apply(get_key, axis=1)
df['Silhouette Score'] = df['silhouette']
df['Davies Bouldin Score'] = df['davies_bouldin']

```

```python
plot_line_chart(
  df_plot=df,
  x='Data Variation',
  y='Silhouette Score',
  title='',
  color='Embedding Model'
)
```

```python
import plotly.graph_objects as go
import numpy as np
```

```python
colors = [
  'DarkBlue',
  'DarkSeaGreen',
  'DarkSlateGray',

  'LightBlue',
  'LightSeaGreen',
  'LightSlateGray',
]

models = list(sorted(df['Embedding Model'].unique()))

def get_scores(metric_name, variation):
  result = np.array([
    df[(df['Embedding Model'] == model) & (df['Data Variation'] == variation)][metric_name].to_list() for model in models
  ]).flatten()

  return result


def plot_bar_chart(metric_name, title):
  fig = px.bar(
    df,
    x='Embedding Model',
    y=metric_name,
    color="Data Variation",
    barmode="group",
    title=title,
    color_discrete_sequence=colors,
  )

  fig.update_layout(
    height=600,
    barmode='group',
  )

  #fig.update_layout(
  #  height=600,
  #  barmode='group',
  #  legend=dict(
  #    orientation="h",
  #    yanchor="bottom",
  #    y=1.02,
  #    xanchor="right",
  #    x=1
  #  ),
  #  font=dict(
  #      size=18
  #  )
  #)

  fig.show()
```
